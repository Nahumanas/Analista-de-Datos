```python
#Importamos Datos 1

#Importar Bibliotecas
import pandas as pd

#Cargamos el archivo CSV
arch = pd.read_csv(r"C:\Users\HP\Desktop\Universidad\Tec CsDatos\Analisis y Visualizacion de Datos\ARCHIVOS\AP2\datos_macro.csv")

#Ver las primeras filas del DataFrame
    #print(arch.head())

#Insepccion de los datos
print(arch.isnull().sum()) 

#Borrar filas duplicadas
arch = arch.drop_duplicates()

#Correccion de tipos de datos 3
#Convertir la columna 'fecha' a tipo datetime
arch['fecha'] = pd.to_datetime(arch['fecha'])

#Convertir la columna 'precio' a tipo float
arch['precio'] = arch['precio'].astype(float)


#Manejo de valores nulos 4 

#Rellenar valores nulos con la media 
arch['cantidad'].fillna(arch['cantidad'].mean(), inplace=True)

#Filtrar datos relevantes 5

#Filtrado 
arch = arch[arch['cantidad'] > 0]

#Validacion Final 6
print(arch.info())



```
